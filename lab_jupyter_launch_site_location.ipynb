```python
## Task 1: Mark all launch sites on a map
import pandas as pd
import io

# Unduh dataset SpaceX
from js import fetch

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Sekarang buat dataframe ringkasan untuk lokasi peluncuran
launch_sites_df = spacex_df[['Launch Site', 'Lat', 'Long']].groupby(['Launch Site'], as_index=False).first()

# Cek apakah datanya sudah masuk
print(launch_sites_df)
# Ambil subset data lokasi peluncuran yang unik
launch_sites_df = spacex_df[['Launch Site', 'Lat', 'Long']].groupby(['Launch Site'], as_index=False).first()

# Inisialisasi peta di koordinat NASA (pusat referensi)
nasa_coordinate = [29.552363, -95.083211]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Iterasi setiap lokasi peluncuran untuk ditambahkan ke peta
for index, site in launch_sites_df.iterrows():
    # Tentukan koordinat lokasi
    coordinate = [site['Lat'], site['Long']]
    
    # 1. Tambahkan Circle (Lingkaran) untuk menandai area lokasi
    circle = folium.Circle(
        coordinate, 
        radius=1000, 
        color='#d35400', 
        fill=True
    ).add_child(folium.Popup(site['Launch Site']))
    
    # 2. Tambahkan Marker dengan label teks menggunakan DivIcon
    marker = folium.map.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12pt; color:#d35400;"><b>%s</b></div>' % site['Launch Site'],
        )
    )
    
    # Masukkan lingkaran dan marker ke dalam objek site_map
    site_map.add_child(circle)
    site_map.add_child(marker)

# Tampilkan peta
site_map
```

```python
# Task 2: Mark the success/failed launches for each site on the map
# 1. Inisialisasi MarkerCluster
marker_cluster = MarkerCluster()

# 2. Buat fungsi untuk menentukan warna marker berdasarkan kolom 'class'
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'  # Hijau untuk Sukses
    else:
        return 'red'    # Merah untuk Gagal

# 3. Tambahkan kolom baru 'marker_color' ke dalam dataframe spacex_df
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# 4. Tambahkan objek marker_cluster ke dalam site_map yang sudah ada
site_map.add_child(marker_cluster)

# 5. Iterasi setiap baris di spacex_df untuk membuat marker di dalam cluster
for index, record in spacex_df.iterrows():
    # Buat objek Marker untuk setiap peluncuran
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        # Gunakan warna yang sudah ditentukan (green/red)
        icon=folium.Icon(color='white', icon_color=record['marker_color'])
    )
    # Masukkan marker ke dalam cluster
    marker_cluster.add_child(marker)

# Tampilkan peta
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # Radius bumi dalam kilometer
    R = 6373.0

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
